### Café Libérica 2.5 kg é o é o com maior margem do lucro
### Café Robusta 0.2 kg é o com menor margem do lucro

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

data_dir = r'./Painel de Vendas de Café.xlsx'
save_path = './images/'
os.makedirs(save_path, exist_ok=True)
produtos = pd.read_excel(data_dir, sheet_name='Produtos')

produtos['Tipo Café'] = produtos['Tipo Café'].replace({
    'Ara': 'Arábica',
    'Rob': 'Robusta',
    'Exc': 'Excelsa',
    'Lib': 'Libérica'
})

produtos['Torra'] = produtos['Torra'].replace({
    'M': 'Média',
    'E': 'Escura',
    'C': 'Clara'
})

produtos_agrupado = produtos.groupby(['Tipo Café', 'Tamanho', 'Torra'])['Lucro'].sum().reset_index()
produtos_agrupado['Lucro'] = produtos_agrupado['Lucro'].round(2)

min_lucro_row = produtos_agrupado.loc[produtos_agrupado['Lucro'].idxmin()]
menor_tipo_cafe = min_lucro_row['Tipo Café']
menor_tamanho = min_lucro_row['Tamanho']
produtos_menor_lucro = produtos_agrupado[(produtos_agrupado['Tipo Café'] == menor_tipo_cafe) & (produtos_agrupado['Tamanho'] == menor_tamanho)]
total_lucro_menor_tipo_tamanho = produtos_menor_lucro['Lucro'].sum()

max_lucro_row = produtos_agrupado.loc[produtos_agrupado['Lucro'].idxmax()]
maior_tipo_cafe = max_lucro_row['Tipo Café']
maior_tamanho = max_lucro_row['Tamanho']
produtos_maior_lucro = produtos_agrupado[(produtos_agrupado['Tipo Café'] == maior_tipo_cafe) & (produtos_agrupado['Tamanho'] == maior_tamanho)]
total_lucro_tipo_tamanho = produtos_maior_lucro['Lucro'].sum()

fig, ax = plt.subplots(figsize=(6, 3))  
ax.axis('tight')
ax.axis('off')

table_data = produtos_menor_lucro.values
columns = produtos_menor_lucro.columns
table = ax.table(cellText=table_data, colLabels=columns, cellLoc='center', loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2) 

plt.title(f'{menor_tipo_cafe} {menor_tamanho} kg (Lucro Total: {total_lucro_menor_tipo_tamanho:.2f})', fontsize=14, weight='bold')

file_name = f'{menor_tipo_cafe}_{menor_tamanho}_menor_lucro_tabela.png'
full_path = os.path.join(save_path, file_name)

plt.savefig(full_path, bbox_inches='tight', pad_inches=0.1)
plt.close()

fig, ax = plt.subplots(figsize=(6, 3))  
ax.axis('tight')
ax.axis('off')

table_data = produtos_maior_lucro.values
columns = produtos_maior_lucro.columns
table = ax.table(cellText=table_data, colLabels=columns, cellLoc='center', loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2) 

plt.title(f'{maior_tipo_cafe} {maior_tamanho} kg (Lucro Total: {total_lucro_tipo_tamanho:.2f})', fontsize=14, weight='bold')

os.makedirs(save_path, exist_ok=True)
file_name = f'{maior_tipo_cafe}_{maior_tamanho}_maior_lucro_tabela.png'
full_path = save_path + file_name

plt.savefig(full_path, bbox_inches='tight', pad_inches=0.1)
plt.close()

lucro_total = produtos_agrupado['Lucro'].sum()

tamanhos = produtos_agrupado['Tamanho'].unique()

for tipo in produtos_agrupado['Tipo Café'].unique():
    for tamanho in tamanhos:
        produtos_tipo_tamanho = produtos_agrupado[(produtos_agrupado['Tipo Café'] == tipo) & (produtos_agrupado['Tamanho'] == tamanho)]
        
        if not produtos_tipo_tamanho.empty:
            total_lucro_tipo_tamanho = produtos_tipo_tamanho['Lucro'].sum()
            
            print(f"\n{tipo} {tamanho} kg (Lucro Total: {total_lucro_tipo_tamanho:.2f}):")
            print(produtos_tipo_tamanho.to_string(index=False))
            
```

### Café Arábica possuí o maior volume de vendas
### 0.5 kg é o tamanho mais vendido de café

```python
encomendas = pd.read_excel(data_dir, sheet_name='Encomendas')

df_agrupado = encomendas.groupby(['Tipo Café Nome', 'Tamanho'])['Quantidade'].sum().reset_index()

total_vendas_por_tipo = encomendas.groupby('Tipo Café Nome')['Quantidade'].sum().reset_index()

tipo_mais_vendido = total_vendas_por_tipo.loc[total_vendas_por_tipo['Quantidade'].idxmax()]
tipo_nome = tipo_mais_vendido['Tipo Café Nome']
total_quantidade = tipo_mais_vendido['Quantidade']


os.makedirs(save_path, exist_ok=True)

print("Tabela de Tipo de Café, Tamnho e Quantidade Vendida:")
for tipo in total_vendas_por_tipo['Tipo Café Nome']:
    df_tipo = df_agrupado[df_agrupado['Tipo Café Nome'] == tipo]
    total_quantidade = total_vendas_por_tipo[total_vendas_por_tipo['Tipo Café Nome'] == tipo]['Quantidade'].values[0]
    
    print(f"\n{tipo} (Total Quantidade Vendida: {total_quantidade}):")
    print(df_tipo.to_string(index=False))

    if tipo == tipo_nome:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.axis('tight')
        ax.axis('off')

        table_data = df_tipo.values
        columns = df_tipo.columns
        table = ax.table(cellText=table_data, colLabels=columns, cellLoc='center', loc='center')

        table.auto_set_font_size(False)
        table.set_fontsize(10), fontsize=14, weight='bold'
        table.scale(1.2, 1.2)

        plt.title(f'{tipo} (Total Quantidade Vendida: {total_quantidade})', fontsize=14, weight='bold')

        file_name = f'{tipo}_maior_venda_tabela.png'
        plt.savefig(full_path, bbox_inches='tight', pad_inches=0.1)
        plt.close()


df_agrupado = encomendas.groupby('Tamanho')['Quantidade'].sum().reset_index()
df_agrupado = df_agrupado.sort_values(by='Quantidade', ascending=False)
df_agrupado['Quantidade'] = df_agrupado['Quantidade'].apply(lambda x: f"{x:.0f}")

print("\nTabela de Tamanho e Quantidade Vendida:")
print(df_agrupado.to_string(index=False))

tamanho_mais_vendido = df_agrupado.loc[df_agrupado['Quantidade'].idxmax()]

fig, ax = plt.subplots(figsize=(6, 2)) 
ax.axis('tight')
ax.axis('off')

table_data = df_agrupado.values
columns = df_agrupado.columns
table = ax.table(cellText=table_data, colLabels=columns, cellLoc='center', loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

plt.title('Tamanhos e Quantidade Vendida', fontsize=14, weight='bold')
file_name = f'{tamanho_mais_vendido['Tamanho']}kg_maior_volume_tabela.png'
plt.savefig(full_path, bbox_inches='tight', pad_inches=0.1)
plt.close()
```

### Café com mais lucro gerado em vendas é Libérica Torra Escura 2.5Kg 

```python
produtos = pd.read_excel(data_dir, sheet_name='Produtos')
encomendas = pd.read_excel(data_dir, sheet_name='Encomendas')

merged_df = pd.merge(encomendas, produtos, on=['Tipo Café', 'Torra', 'Tamanho'], how='left')

merged_df['Lucro Gerado'] = merged_df['Quantidade'] * merged_df['Lucro']

resultado = merged_df.groupby(['Tipo Café Nome', 'Torra', 'Tamanho'])[['Quantidade', 'Lucro Gerado']].sum().reset_index()

torra_renaming = {'E': 'Escura', 'C': 'Clara', 'M': 'Média'}
resultado['Torra'] = resultado['Torra'].replace(torra_renaming)

resultado = resultado.sort_values(by='Lucro Gerado', ascending=False).round(2)
top_5 = resultado.head(5)

print(top_5.to_string(index=False))

fig, ax = plt.subplots(figsize=(12, 6))
ax.axis('tight')
ax.axis('off')

table_data = top_5.values
columns = top_5.columns
table = ax.table(cellText=table_data, colLabels=columns, cellLoc='center', loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2) 

plt.title('Top 5 Produtos com Maior Lucro Gerado', fontsize=14, weight='bold')

os.makedirs(save_path, exist_ok=True)
file_name = 'resultado_tabela.png'

plt.savefig(full_path, bbox_inches='tight', pad_inches=0.1)
plt.close()
```
